```python
def load(ym, ref):
    dvc_path = f"s3/ctbk/normalized/{ym}.parquet.dvc"
    blob = repo.commit(ref).tree[dvc_path]
    # print(blob)
    dvc_spec = yaml.safe_load(blob.data_stream)
    out = singleton(dvc_spec['outs'], dedupe=False)
    md5 = out['md5']
    cache_path = relpath(join(repo.working_dir, '.dvc', 'cache', 'files', 'md5', md5[:2], md5[2:]))
    df = read_parquet(cache_path)
    return df
```
